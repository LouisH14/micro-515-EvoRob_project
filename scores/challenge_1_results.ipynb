```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy

import scienceplots

plt.style.use(["science", "nature"])
```

```python
results = [
    {"team_id": "A", "team_name": "GROUPA", "score": -1285.21},
    {"team_id": "B", "team_name": "groupB", "score": 1742.0},
    {"team_id": "C", "team_name": "The Comeback Team", "score": 260.0},
    {"team_id": "D", "team_name": "SPIDERMAN", "score": -559.33},
    {"team_id": "E", "team_name": "Gp-E", "score": 3517.47},
    {"team_id": "F", "team_name": "Group F", "score": 1181.63},
    {"team_id": "G", "team_name": "GTEAM", "score": 457.11},
    {"team_id": "H", "team_name": "Exodia", "score": 4392.45},
    {"team_id": "I", "team_name": "4LegsIsNotEnough", "score": 1048.12},
    {"team_id": "W", "team_name": "groupW", "score": 1560.11},
]

solution = {"score": 1432.21}
```

```python
df = pd.DataFrame(results).sort_values("score", ascending=True)
df["label"] = df["team_id"] + " - " + df["team_name"]
solution_score = float(solution["score"])

plt.figure(figsize=(8.5, 4))
colors = plt.cm.tab10(numpy.linspace(0, 1, len(df)))
bars = plt.barh(df["label"], df["score"], color=colors)
plt.axvline(0, color="black", linewidth=1)
plt.axvline(solution_score, color="crimson", linestyle="--", linewidth=1.5, label=f"Solution: {solution_score}")
plt.title("Challenge 1 Team Scores")
plt.xlabel("Score")
plt.ylabel("Team")

# Annotate each bar with its numeric score for readability.
for bar, value in zip(bars, df["score"]):
    x_pos = value + 30 if value >= 0 else value - 220
    plt.text(x_pos, bar.get_y() + bar.get_height() / 2, f"{value:.0f}", va="center")

y_top = len(df) - 10
plt.text(solution_score + 32, y_top, f"TA solution: {solution_score:.0f}", color="crimson", va="bottom", fontsize=9)

plt.tight_layout()
plt.savefig("challenge_1_scores.svg", dpi=300, transparent=True)
plt.show()

df.sort_values("score", ascending=False).reset_index(drop=True)
```
